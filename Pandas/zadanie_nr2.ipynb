## Import bibliotek

```python
import numpy as np
import pandas as pd
```

## Wczytywanie danych

```python
def fetch_financial_data(company=''):
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('META')
df.head()
```

## Zapis/odczyt danych z pliku CSV

### Zapis

```python
df.to_csv('dane_meta.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2024)]
df_nov
```

```python
df_nov.to_csv('dane_meta_nov.csv')
```

```python
df_nov.to_excel('dane_meta_nov.xlsx')
```
