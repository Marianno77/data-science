Wycinanie fragmentów obrazu

```python
img = cv2.imread('img1.jpg')
cv2_imshow(img)
```

```python
left_top_corner = img[:100, :100]
cv2_imshow(left_top_corner)
```

```python
right_top_corner = img[:100, -100:]
cv2_imshow(right_top_corner)
```

```python
left_bottom_corner = img[-100:, :100]
cv2_imshow(left_bottom_corner)
```

```python
rigth_bottom_corner = img[-100:, -100:]
cv2_imshow(rigth_bottom_corner)
```

```python
height, width, _ = img.shape
h0 = height//2
w0 = width//2
center = img[h0-50:h0+50, w0-50:w0+50]
cv2_imshow(center)
```

## Zmiana rozmiaru obrazu

```python
img = cv2.imread('img2.jpg')
img_resized = imutils.resize(img, height=400)
cv2_imshow(img)
print(img.shape)
cv2_imshow(img_resized)
print(img_resized.shape)
```

```python
img_resized = imutils.resize(img, width=400)
cv2_imshow(img_resized)
print(img_resized.shape)
```

## Detekcja krawędzi

```python
cv2_imshow(cv2.Laplacian(img, ddepth=cv2.CV_64F))
```

```python
cv2_imshow(cv2.Canny(img, threshold1=100, threshold2=200))
```

## Rysowanie po obrazie

```python
cv2_imshow(cv2.rectangle(img.copy(), pt1=(60,50), pt2=(115,100), color=(0,255,0), thickness=2))
```

```python
cv2_imshow(cv2.circle(img.copy(), center=(145,30), radius=30, color=(0,255,0), thickness=2))
```

```python
cv2_imshow(cv2.circle(img.copy(), center=(145,30), radius=30, color=(0,255,0), thickness=-1))
```

```python
cv2_imshow(cv2.putText(
                        img.copy(),
                        text='IMAGE',
                        org=(10,30),
                        fontFace=cv2.FONT_HERSHEY_DUPLEX,
                        fontScale=1,
                        color=(0,255,0),
                        thickness=2))
```
