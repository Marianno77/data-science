# **Matplotlib**

## Import bibliotek

```python
import matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

## Wykres Liniowy

```python
_ = plt.plot([0, 1], [0, 5])
```

```python
plt.plot([0, 1], [0, 5], label = 'Line' , color = 'orange')
_ = plt.legend()
```

```python
plt.plot([0, 1 , 2], [0, 5, 2], label = 'Line' , color = 'orange')
plt.xlabel('OX')
plt.ylabel('OY')
plt.title('Title')
_ = plt.legend()
```

```python
plt.plot([0, 1], [0, 5], label = 'Line1' , color = 'orange')
plt.plot([1, 2], [5, 2], label = 'Line2' , color = 'purple')
plt.plot([1, 2], [0, 5], label = 'Line3' , color = 'blue')
plt.xlabel('OX')
plt.ylabel('OY')
plt.title('Title')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

## Tworzenie wielu wykresów

```python
plt.subplot(1, 2, 1)
plt.plot([0, 1], [0, 1], color = 'orange')

plt.subplot(1, 2, 2)
plt.plot([0, 1], [0, 1], 'r--')
```

```python
fig = plt.figure(figsize = (10, 5))

plt.subplot(1, 3, 1)
plt.plot([0, 1], [0, 1], color = 'orange')

plt.subplot(1, 3, 2)
plt.plot([0, 1], [0, 1], 'r--')

plt.subplot(1, 3, 3)
_ = plt.plot([0, 1], [0, 1], 'go')
```


## Wykresy słupkowe

```python
_ = plt.bar(x=[0,1,2], height=[1,2,3])
```

```python
_ = plt.bar(x=['A','B','C'], height=[1,2,3], color=['red','green','blue'])
```

```python
_ = plt.barh(y=['A','B','C'], width=[1,2,3], color=['red','green','blue'])
```

## Wykresy punktowe

```python
x = np.random.rand(10)
y = np.random.rand(10)

_ = plt.scatter(x, y)
```

```python
size = np.random.rand(10)*500
_ = plt.scatter(x, y, s=size)
```

```python
_ = plt.scatter(x, y, s=size, alpha=0.5)
```
