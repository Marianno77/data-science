# **Matplotlib - style**

## Import bibliotek

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)

plt.plot(x,x_sin)
plt.show()
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
plt.plot(x,x_sin)
plt.show()
```

```python
plt.style.use('seaborn-v0_8-ticks')
plt.plot(x,x_sin)
plt.show()
```

```python
plt.style.use('dark_background')
plt.plot(x,x_sin)
plt.show()
```

```python
plt.style.use('seaborn-v0_8')
plt.plot(x,x_sin)
plt.show()
```
