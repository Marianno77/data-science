# **Regresja liniowa**

## Import bibliotek

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import sklearn

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#from sklearn.metrics import mean_squared_error, r2_score
```

## Wygemerowanie danych

```python
x = np.arange(0, 50, 0.5)
noise = 10 * np.random.randn(100)
y = 2 * x + 100 + noise
x = x.reshape(-1, 1)

print(x.shape)
print(y.shape)
```

## Poddział danych na zbiór treningowy i testowy

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Wizualizacja

```python
plt.scatter(x_train, y_train, label='Treningowy')
plt.scatter(x_test, y_test, label='Testowy')
_ = plt.legend()
```

## Regresja liniowa

```python
  lin_reg = LinearRegression()
  lin_reg.fit(x_train, y_train)

  print(lin_reg.coef_)
  print(lin_reg.intercept_)
```

## Wizualizacja zbioru treningowego i dopasowanego modelu

```python
plt.title('Regresja liniowa - jedna zmienna')
plt.scatter(x_train, y_train, label='Treningowy')
plt.plot(x, lin_reg.intercept_ + lin_reg.coef_[0] * x, c='red')
_ = plt.legend()
```

## Wizualizacja zbioru testowego i dopasowanego modelu

```python
y_pred = lin_reg.predict(x_test)
plt.title('Regresja liniowa - jedna zmienna')
plt.scatter(x_test, y_test, label='Testowy')
plt.plot(x, lin_reg.intercept_ + lin_reg.coef_[0] * x, c='red')
_ = plt.legend()
```

## Ocena modelu

```python
lin_reg.score(x_test, y_test)
```

## Przykład złego zastosowania regresji liniowej

```python
x = np.arange(-5, 5, 0.5)
noise = 10 * np.random.randn(20)
y = 2 * (x ** 2) + 4 + noise
x = x.reshape(-1, 1)

lin_reg = LinearRegression()
lin_reg.fit(x, y)

print(lin_reg.coef_)
print(lin_reg.intercept_)

plt.title('Regresja liniowa - jedna zmienna')
plt.scatter(x, y, label='Dane')
plt.plot(x, lin_reg.intercept_ + lin_reg.coef_[0] * x, c='red')
_ = plt.legend()
```
